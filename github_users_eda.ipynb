```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: API URL
url = "https://api.github.com/users?per_page=50"

# Step 2: Fetch data
response = requests.get(url)
data = response.json()
```

```python
# we converted it into data frame
df = pd.DataFrame(data)
```

```python
df
```

```python
# basic EDA
print("shape", df.shape) #checking dataset shape
```

```python
print("Missing values:\n", df.isnull().sum()) # it means there is no any missing value
```

```python
#check the duplicate 
print("duplicate:", df.duplicated().sum()) #there is no any duplicate values
```

```python
# unique user type (user/organization)
print("\nUnique types:", df['type'].unique()) #total 2 unique categories
```

```python
df = df[['login','id','type','site_admin']] #selecting real useful column
df
```

###  analysing the data

```python
# counting different user types
print(df['type'].value_counts())
```

```python
print(df['site_admin'].value_counts()) # checking number of site admins
```

```python
# adding new feature
df['login_length'] = df['login'].apply(len)
print(df.head())
```

```python
#checking relationship between id and loginlength
print(df[['id', 'login_length']].corr())
```

```python
print(df.groupby('type')['login_length'].mean()) #here is average of types (organization, user)
```

```python
print("top 10 longest user name:")
print(df.sort_values('login_length', ascending=False).head(10)) # Finding top 10 longest usernames
```

```python
#re naming the column name
df.rename(columns={'login': 'login_name'}, inplace=True)
```

```python
df
```

```python
#Average username length
print(df['login_length'].mean())
```

### Visualization

```python
plt.tight_layout()
plt.style.use('ggplot')
```

```python
plt.figure(figsize=(6,4))
df['type'].value_counts().plot(kind='bar') #user type distribution
plt.title("Github user types")
plt.xlabel("type")
plt.ylabel("Count")
plt.show()
```

```python
df['login_length'].plot(kind='hist', bins = 10) #Username length distribution
plt.title("username length distribution")
plt.xlabel("length")
plt.ylabel("frequency")
plt.show()
```

```python
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%') #user type percentage distribution
plt.title("User Type Distribution")
plt.ylabel("")
plt.show()
```
